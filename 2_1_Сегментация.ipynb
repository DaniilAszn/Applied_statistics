## Этап 2: Сегментация.

Будем сегментировать клиентов по RFM.
* **Recency (Свежесть)** - число дней насколько давно пользователь что-то у нас покупал.
* **Frequency (Частота)** - насколько много заказов пользователь сделал.
* **Monetary (Деньги)** - сколько денег в суммме потратил.

Клиенты, которые пришли недавно, много и часто покупали - лучшие клиенты.

```python
uk_data = uk_data[['CustomerID', 'InvoiceDate', 'InvoiceNo', 'Quantity', 'UnitPrice']]
uk_data.head()
```

```python
# колонка с итоговой суммой, потраченной при покупке
uk_data['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
# диапазон, когда покупали заказы
uk_data['InvoiceDate'].min(), uk_data['InvoiceDate'].max()
```

```python
import datetime as dt

PRESENT = dt.datetime(2011, 12, 10) # самая новая дата

uk_data['InvoiceDate'] = pd.to_datetime(uk_data['InvoiceDate'])
```

```python
rfm = (
    uk_data.groupby('CustomerID')
    .agg({
        'InvoiceDate': lambda date: (PRESENT - date.max()).days,
        'InvoiceNo': 'count',
        'TotalPrice': 'sum'
    })
)
```

```python
rfm.columns
```

```python
rfm.columns = ['recency', 'frequency', 'monetary']
```

```python
rfm.head()
```

```python
rfm[['recency', 'frequency', 'monetary']].hist(figsize=(12,6), bins=30);
```

Видим довольно длинные хвосты. Для более информативной визуализации прологарифмируем наши переменные.

```python
g = (
    rfm[['recency', 'frequency', 'monetary']]
    .apply(lambda w: np.log(w + 1))
    .hist(figsize=(12,6), bins=30)
)

g;
```

```python
rfm[['recency', 'frequency', 'monetary']] = rfm[['recency', 'frequency', 'monetary']].apply(lambda w: np.log(w + 1))
```

Мы будем использовать для сегментации квартили, поэтому для нас выбросы не проблема. Данные очищать не будем.
Номер соответствующего квартиля помогает найти команда `qcut()`

```python
rfm['r_quantile'] = pd.qcut(rfm['recency'], 4, [str(i) for i in range(1, 5)])
rfm['f_quantile'] = pd.qcut(rfm['frequency'], 4, [str(i) for i in reversed(range(1, 5))])
rfm['m_quantile'] = pd.qcut(rfm['monetary'], 4, [str(i) for i in reversed(range(1, 5))])
```

```python
rfm.head()
```

Создадим единую колонку с номерами сегментов.

```python
rfm['RFM_Score'] = rfm.r_quantile.astype(str) + rfm.f_quantile.astype(str) + rfm.m_quantile.astype(str)
```

```python
rfm
```

Посмотрим в какой сегмент сколько пользователей попало. Нас больше всего интересует сегмент 111. Меньше всего 444.

```python
rfm['RFM_Score'].value_counts()[:10]
```

Выберем из таблички наших самых преданных покупателей.

```python
rfm[rfm['RFM_Score'] == '111'].sort_values('monetary', ascending=False).head()
```

Для визуализации оставим только те кластеры, куда попало больше 100 человек.

```python
clusters = rfm['RFM_Score'].value_counts()
big_clusters = clusters[clusters > 100].index.values
big_clusters
```

Нарисуем интерактивную визуализацию.

```python
import plotly.express as px

px.scatter_3d(
    rfm[rfm.RFM_Score.isin(big_clusters)],
    x='recency',
    y='frequency',
    z='monetary',
    color='RFM_Score'
)
```
