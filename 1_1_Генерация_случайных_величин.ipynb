# Симуляции случайных величин

```python
import numpy as np      # математика
import pandas as pd     # таблицы
from scipy import stats # статистика

#визуализация
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
%matplotlib inline
```

## 1. Генерация случайных величин

```python
norm_rv = stats.norm(loc=0, scale=1) # задали генератор
```

```python
sample = norm_rv.rvs(1000) # выборка из 1000 значений
sample[:5]
```

## 2. Плотность и функция  распределения

Плотность распределения

```python
norm_rv.pdf(1) # f(1)
```

```python
x = np.linspace(-3, 3, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.scatter([1], [norm_rv.pdf(1)], color='blue')

# доп. код для отрисовки заштрихованой линии
xq = np.linspace(-3, 1)
yq = norm_rv.pdf(xq)

plt.fill_between(xq, 0, yq, color='blue', alpha=0.5)
plt.axvline(1, color='blue', linestyle='--', lw=2)
```

Функция распределения

```python
norm_rv.cdf(1)
```

```python
x = np.linspace(-3, 3, 100)
cdf = norm_rv.cdf(x)

plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')

plt.scatter([1], [norm_rv.cdf(1)], color='blue')

# доп. код для отрисовки заштрихованой линии
xq = np.linspace(-3, 1)
yq = norm_rv.cdf(xq)

plt.fill_between(xq, 0, yq, color='blue', alpha=0.5)
plt.axvline(1, color='blue', linestyle='--', lw=2)
```

```python
print(norm_rv.cdf(2.5) - norm_rv.cdf(1))
```

```python
x = np.linspace(-3, 3, 100)
cdf = norm_rv.cdf(x)

plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')

plt.scatter([1], [norm_rv.cdf(1)], color='blue')
plt.scatter([2.5], [norm_rv.cdf(2.5)], color='blue')

# доп. код для отрисовки заштрихованой линии
xq = np.linspace(1, 2.5)
yq = norm_rv.cdf(xq)

plt.fill_between(xq, 0, yq, color='blue', alpha=0.5)
plt.axvline(1, color='blue', linestyle='--', lw=2)
plt.axvline(2.5, color='blue', linestyle='--', lw=2)
```

Квантиль - такое число q, что P(X<=q) = gamma

```python
q = norm_rv.ppf(0.1)
q
```

```python
x = np.linspace(-3, 3, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

xq = np.linspace(-3, q)
yq = norm_rv.pdf(xq)

plt.fill_between(xq, 0, yq, color='blue', alpha=0.5)
plt.axvline(q, color='blue', linestyle='--', lw=2)

y_max = plt.ylim()[1]
plt.text(q + 0.1, 0.8 * y_max, round(q, 2), color='blue', fontsize=16)
```

Экспоненциальное распределение

```python
exp_rv = stats.expon(scale=0.5)
exp_rv.rvs(5)
```

```python
x = np.linspace(-3, 3, 100)
pdf = norm_rv.pdf(x)

# плотность
plt.plot(x, pdf, lw=3)

# гистограмма
plt.hist(sample, bins=30, density=True)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Построение эмпирической функции распределения по выборке

```python
from statsmodels.distributions.empirical_distribution import ECDF

ecdf = ECDF(sample)

plt.step(ecdf.x, ecdf.y)
plt.ylabel('$F(x)$',fontsize=20)
plt.xlabel('$x$', fontsize=20)
```

```python
# Эмпирическая функция
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='Empirical CDF')

# Теоретическая функция
x = np.linspace(-3, 3, 100)
cdf = norm_rv.cdf(x)
plt.plot(x, cdf, label='Theoretical CDF')

plt.ylabel('$F(x)$')
plt.xlabel('$x$')

plt.legend(loc='upper left')
```

## 5. Зёрна генерации

без фиксации:

```python
norm = stats.norm(loc=3, scale=2)
norm.rvs(5)
```

```python
norm = stats.norm(loc=3, scale=2)
norm.rvs(5)
```
